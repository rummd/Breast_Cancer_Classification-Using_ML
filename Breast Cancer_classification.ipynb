```python
import pandas as pd
import numpy as np
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Collecting data and processing

```python
# loading the data

# converting numpy array data to pandas data frame
# data_frame= pd.DataFrame(breast_cancer_dataset.data, column=breast_cancer_dataset.feature_names)



# pandas
data_frame=pd.read_csv('data.csv')
data_frame.head(8)
```

```python
# assigning 1/0 for diagnosis to the data frame
data_frame['diagnosis'] = data_frame['diagnosis'].map({"M":0,"B":1})
```

```python
# dropping unnecessary columns/rows
data_frame.drop(columns = ['id','Unnamed: 32'],inplace = True)
data_frame.info()
```

```python
#  SEPARATING THE features and target
```

```python
X=data_frame.drop(columns='diagnosis', axis=1)
Y=data_frame['diagnosis']
```

```python
X
```

```python
Y
```

# Splitting the data into training and testing sets

```python
X_train,X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Model Training
Logistic Regression

```python
model=LogisticRegression()
```

```python
#training the model using training data
model.fit(X_train,Y_train)
```

Model Evaluation


Accuracy Score

```python
# accuracy in training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)
```

```python
print('accuracy on training data=',training_data_accuracy)
```

```python
# accuracy in test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python
print('accuracy on test data=',test_data_accuracy)
```
